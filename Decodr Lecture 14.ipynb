```python
sns.set_style("whitegrid")
category_order=["Never","Rarely","Sometimes",
               "often","Always"]
sns.catplot(x="Parents Advice",data=survey_data,
           kind="count",order=category_order)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("Purples")
category_order=["Never","Rarely","Sometimes",
               "often","Always"]
sns.catplot(x="Parents Advice",data=survey_data,
           kind="count",order=category_order)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("RdBu")
category_order=["Never","Rarely","Sometimes",
               "often","Always"]
sns.catplot(x="Parents Advice",data=survey_data,
           kind="count",order=category_order)
plt.show()
```
